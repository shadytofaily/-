### Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates

```python
a =  np.arange(20,dtype = float)
cart = np.reshape(a,(10,2))
x = cart[:,0]
y = cart[:,1]
rho = np.sqrt(x**2 + y**2 )
phi = np.arctan2(y, x) 
polar = np.empty((10,2))
polar[:,0] = rho
polar[:,1] = phi
polar
```

# Task 3
## $Snail$ $curve$

```python
phi = np.linspace(-2*np.pi, 2*np.pi, 201)

a = np.linspace(-5, 5, 22)

import matplotlib.pyplot as plt
for each in a:
    r = np.cos(phi) + each
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    plt.plot(x, y)
plt.show()
```

# Task 4
## $Cochleoid$  $curve$

```python
phi = np.linspace(-2*np.pi, 2*np.pi, 201)

phi = phi[np.where(phi!=0.)]

a = np.linspace(-5, 5, 22)

import matplotlib.pyplot as plt
for each in a:
    r =  each * np.sin(phi) / phi
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    plt.plot(x, y)
plt.show()
```

## Construct the Cauchy matrix using:  
### 1) Explicit broadcasting

```python
u = np.arange(4)
v = u + 0.5
np.set_printoptions(precision=3)
br = 1 / (u[:, None] - v)
print(br)
```

### 2) subtract.outer attribute

```python
c = 1 / np.subtract.outer(u, v)
c
```
