```python
files = ['test_00.json', 'test_01.json', 'test_02.json', 'test_03.json', 'test_04.json', 'test_05.json', 'test_06.json', 'test_07.json', 'test_08.json', 'test_09.json']
texts = []
answers = []
targets = []
f = []
for filename in files:
    file = pd.read_json(f'{filename}')
    f.append(file)
```

```python
# here making a list of all tasks from all the files
tasks = []
for i in range(len(f)):
    task = f[i]['tasks']
    for j in range(len(task[i])):
        task = f[i]['tasks'][j]
        tasks.append(task)
```

```python
# filtering the tasks,which we need
needed_tasks = []
for i in range(len(tasks)):
    if ('ГЛАВНАЯ' or 'главная') in tasks[i]['text']:
        needed_tasks.append(tasks[i])
```

```python
zad_txt =[]
otv_txt= []
solutions = []
otv_ch = []
targets = []
task_nb = []
for i in range(len(needed_tasks)):
    for j in range(len(needed_tasks[i]['question']['choices'])):
        otv_txt.append(needed_tasks[i]['question']['choices'][j]['text'])
        zad_txt.append(needed_tasks[i]['text'])
        target = needed_tasks[i]['solution']['correct_variants'][0]
        solutions.append(target)
        targets.append(1 if needed_tasks[i]['question']['choices'][j]['id'] in target else 0)
        task_nb.append(i+1)
```

```python
for i in range(len(zad_txt)):
    f = zad_txt[i].find('(')
    zad_txt[i] =zad_txt[i][f:]
```

```python
task_series = pd.Series(task_nb)
target_series = pd.Series(targets)
#sol_series = pd.Series(solutions)
zad_series = pd.Series(zad_txt)
otv_series = pd.Series(otv_txt)
frame = {'Task №':task_series ,'Task text': zad_series, 'Answers': otv_series, 'Target':target_series}
result_df=pd.DataFrame(frame)
result_df
```

```python
pasti = 0 
l_answer = []
l_questions = []
for each in result_df.itertuples(): # each is a tuple with 5 values: index,task nb,text,answer and target
    ind = each[1]
    question = each[2]
    ans = each[3]
    q_words = question.split(' ') # make a list from text 
    ans_words = ans.split(' ')   # list for answers
    ans_words_with_ind = [ind,ans_words]
    if pasti == ind:
        l_answer.append(ans_words_with_ind) 
    else:
        l_answer.append(ans_words_with_ind) 
        l_questions.append(q_words)
    pasti = ind
```

```python
corrected_q =[] # очистка от дополнительных знаков 
for i in range(len(l_questions)):
    question = l_questions[i]
    q = []
    for each in question:
        if '1' in each:
            each = each.replace('1','')
        if '2' in each:
            each = each.replace('2','')
        if '3' in each:
            each = each.replace('3','')
        if ',' in each:
            each = each.replace(',','')
        if '.' in each:
            each = each.replace('.','')
        if '«' in each:
            each = each.replace('«','')
        if '»' in each:
            each = each.replace('»','')
        if '(' in each:
            each = each.replace('(','')
        if ')' in each:
            each = each.replace(')','')
        if '”' in each:
            each = each.replace('”','')
        if '"' in each:
            each = each.replace('"','')
        if '“' in each:
            each = each.replace('“','')            
        if '\xa01' in each:
            each = each.replace('\xa01',' ')
            pos = each.find(' ')
            each = each[:pos]
        if '<' in each:
            each = each.replace('<','')
        if '>' in each:
            each = each.replace('>','')
        each = each.lower()
        if len(each) >= 4:
            q.append(each)
    corrected_q.append(q)
```

```python
corrected_a =[] # очистка от дополнительных знаков 
for i in range(len(l_answer)):
    answer = l_answer[i][1]
    a = []
    for each in answer:
        if '1' in each:
            each = each.replace('1','')
        if '2' in each:
            each = each.replace('2','')
        if '3' in each:
            each = each.replace('3','')
        if ',' in each:
            each = each.replace(',','')
        if '.' in each:
            each = each.replace('.','')
        if '«' in each:
            each = each.replace('«','')
        if '»' in each:
            each = each.replace('»','')
        if '(' in each:
            each = each.replace('(','')
        if ')' in each:
            each = each.replace(')','')
        if '”' in each:
            each = each.replace('”','')
        if '"' in each:
            each = each.replace('"','')
        if '“' in each:
            each = each.replace('“','')            
        if '\xa01' in each:
            each = each.replace('\xa01',' ')
            pos = each.find(' ')
            each = each[:pos]
        if '<' in each:
            each = each.replace('<','')
        if '>' in each:
            each = each.replace('>','')
        each = each.lower()
        if len(each) >= 4:
            a.append(each)
    a =[i//5 +1, a]
    corrected_a.append(a)
```

```python
ans_words = corrected_a
text_words = corrected_q
```

```python
differences = []
intersections = []

for i in range(len(ans_words)):
    diff = 0
    inter = 0
    for word in ans_words[i][1]:
        if word in text_words[ans_words[i][0]-1]:
            inter +=1
        else:
            diff += 1
    intersections.append(inter)
    differences.append(diff)
    
col1 = pd.Series(intersections)
col2 = pd.Series(differences)
```

```python
from sklearn import preprocessing
le_q = preprocessing.LabelEncoder()
le_a = preprocessing.LabelEncoder()

list_questions =[]
list_answers =[]

for i in range(40):
    one_ans =result_df['Answers'][i]
    list_answers.append(one_ans)

for i in range(8):
    one_ques =result_df['Task text'][5*i]
    list_questions.append(one_ques)

le_a.fit(list_answers)
le_q.fit(list_questions)

len(list_answers),len(list_questions)
tr_q = le_q.transform(result_df['Task text'])
tr_a = le_a.transform(result_df['Answers'])

result_df['Task text'] = tr_q
result_df['Answers'] = tr_a
```

```python
X = result_df.drop(['Target','Task text','Answers','Task №'],axis =1)
y = result_df.Target
```

```python
import numpy as np
from sklearn.model_selection import GroupKFold
groups = []
for i in range(1,9):
    groups += [i]*5
groups = np.array(groups)
group_kfold = GroupKFold(n_splits=8)
```

```python
groups
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100,random_state=0)
from sklearn.model_selection import cross_val_score
```

```python
n = cross_val_score(clf, X, y ,groups = groups,cv = group_kfold)
```

```python
print(n)
print('Mean =',np.mean(n),'; Max = ',np.max(n))#Min = ',np.min(n))
# results of cross validation
```

```python
# trying to predict, splitting into train and test 
for train_index, test_index in group_kfold.split(X, y, groups):
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
clf.fit(X_train,y_train)
clf.predict(X_test) 
```

```python
from sklearn.metrics import roc_auc_score
print('Train score = ',roc_auc_score(y_train, clf.predict(X_train)))
print('Test score = ',roc_auc_score(y_test, clf.predict(X_test)))
```
